```python
# 從 Google Colab 匯入 userdata（用來安全存放敏感資料，例如 token）
from google.colab import userdata


# 從 Colab 的使用者密鑰中取得 NGROK_AUTHTOKEN
# 這個 token 是你在 ngrok 官網註冊後拿到的，用來啟用 ngrok 服務
ngrok_authtoken = userdata.get('NGROK_AUTHTOKEN')


# 設定本機伺服器要使用的埠號（port）
# Flask 等後端通常會跑在這個 port，例如 http://localhost:5051
port = 5051
```

```python
# 匯入 os 模組（用來操作系統環境，例如環境變數、檔案路徑）
import os

# 從 pyngrok 套件匯入 ngrok（用來建立公開網址，讓外部可以連到本機服務）
from pyngrok import ngrok
```

```python
# 關閉所有目前正在運行的 ngrok 連線（tunnel）
ngrok.kill()
```

```python
# 設定 ngrok 的驗證 token（來自 ngrok 官網註冊）
# 這樣可以啟用你的帳號功能，例如固定網址、更多連線數等
ngrok.set_auth_token(ngrok_authtoken)


# 建立一個 ngrok tunnel（把本機的 port 對外公開）
# port：你的 Flask 伺服器執行的埠號（例如 5051）
# name：自訂這個 tunnel 的名稱（方便辨識）
tunnel = ngrok.connect(port, name="linebot_tunnel")


# 印出公開網址（這個網址要貼到 Line Developers 的 webhook）
print(tunnel.public_url)
```

```python
# 匯入 Flask 框架（用來建立 Web 伺服器）
from flask import Flask


# 建立 Flask 應用程式實例
# __name__ 代表目前這個 Python 檔案
app = Flask(__name__)


# 設定路由（網址路徑）
# 當使用者進入首頁 "/" 時，會執行 home() 這個函式
@app.route('/')
def home():
    # 回傳 HTML 內容（瀏覽器會顯示這段文字）
    return "<h1>明新科技大學 B12090096</h1>"


# 主程式入口（確保這段只在直接執行時才會跑）
if __name__ == '__main__':
    # 啟動 Flask 伺服器，監聽 port 5000
    app.run(port=5000)
```

```python
# 關閉所有目前運行中的 ngrok tunnel（公開連線）
# 常用在重新建立連線前，避免重複開多個網址
ngrok.kill()
```
